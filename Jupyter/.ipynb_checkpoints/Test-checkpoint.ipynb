```python
import numpy as np
from matplotlib import pyplot as plt
```

Consider the following equation:

$$f(x) = -2x^2 + 28x - 97$$

```python
x_vals = np.arange(-5, 20)
y_vals = -2 * (x_vals)**2 + 28 * x_vals - 97

plt.title('Transformations')
plt.plot(x_vals, y_vals)
plt.grid(True)
```
